```python
import numpy as np
import time

# NumPyを使った行列計算
start_time = time.time()

A_np = np.array([[1, 2], [3, 4]])
B_np = np.array([[5, 6], [7, 8]])

C_np = np.dot(A_np, B_np)

end_time = time.time()

print("NumPy:")
print(C_np)
print("--- %s seconds ---\n" % (end_time - start_time))


# Pythonのリストとforループを使った行列計算
start_time = time.time()

A_list = [[1, 2], [3, 4]]
B_list = [[5, 6], [7, 8]]

C_list = [[0, 0], [0, 0]]
for i in range(len(A_list)):
    for j in range(len(B_list[0])):
        for k in range(len(B_list)):
            C_list[i][j] += A_list[i][k] * B_list[k][j]

end_time = time.time()

print("Python list:")
print(C_list)
print("--- %s seconds ---\n" % (end_time - start_time))
```

```python
#特異値分解を使って画像データをU, Σ, V^Tの3つの行列に分解し、
#その中で大きな寄与を持つ特異値のみを取り出して画像を再構成することで、
#元の画像に近い画像をより少ないデータ量で表現する

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
import japanize_matplotlib

# サンプル画像の読み込み
image = io.imread('/content/sakurajima.jpeg')
# カラー画像をグレースケールに変換
gray_image = color.rgb2gray(image)

# グレースケール画像の表示
plt.imshow(gray_image, cmap='gray')
plt.title('圧縮前')
plt.axis('off')
plt.show()

# 画像データの特異値分解
U, S, VT = np.linalg.svd(gray_image, full_matrices=False)

# 特異値の数を指定
k = 50

# k個の特異値を使って画像を再構成
compressed_image = np.dot(U[:, :k], np.dot(np.diag(S[:k]), VT[:k, :]))

# 圧縮後の画像の表示
plt.imshow(compressed_image, cmap='gray')
plt.title(f'圧縮後 (特異値の数={k})')
plt.axis('off')
plt.show()
```

```python
# 3x3の値を生成
val = np.arange(0, 9).reshape(3, 3)

# DataFrameの作成
df = pd.DataFrame(val, index=list('def'), columns=list('abc'))


# NaNを設定
df.loc['d', 'a'], df.loc['e', 'b'] = np.nan, np.nan

# すべてのNaNを0で置き換え
df_zero_filled = df.fillna(value=0)

# カラムごとに異なる値でNaNを置き換え
filling = {'a': 100, 'b': 200, 'c': 300}
df_custom_filled = df.fillna(value=filling)

df_zero_filled, df_custom_filled
```

```python
# 5x5の値を生成
val = np.arange(0, 25).reshape(5, 5)

# DataFrameの作成
df = pd.DataFrame(val, columns=list('abcde'))

# 対角線上の要素をNaNに設定
for i, c in enumerate(df.columns):
    df.loc[i, c] = np.nan

# 'b' 列に対して線形補間
b_interpolated = df['b'].interpolate(method='linear')

# DataFrame全体に対して線形補間（列に沿って）
df_interpolated = df.interpolate(method='linear', axis=1)

b_interpolated, df_interpolated
```
