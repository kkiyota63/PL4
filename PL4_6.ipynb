```python
#このデータセットには、脳腫瘍画像から抽出された特徴を表すデータが含まれている。
#Image column defines image name and Class column defines either the image has tumor or not (1 = Tumor, 0 = Non-Tumor)
brain_tumor_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Brain Tumor.csv")
brain_tumor_data
```

```python
from scipy import stats

group0 = brain_tumor_data[brain_tumor_data['Class'] == 0]['Mean']
group1 = brain_tumor_data[brain_tumor_data['Class'] == 1]['Mean']

_,t_test_result = stats.ttest_ind(group0, group1, equal_var=False)

print("p値は",t_test_result)
```

```python
# prompt: group0, group1の平均、標準偏差、分散を計算

print("非脳腫瘍患者の平均ピクセルの平均:", group0.mean())
print("脳腫瘍患者の平均ピクセルの平均:", group1.mean())
print("")
print("非脳腫瘍患者の平均ピクセルの標準偏差:", group0.std())
print("脳腫瘍患者の平均ピクセルの標準偏差:", group1.std())
print("")
print("非脳腫瘍患者の平均ピクセルの分散:", group0.var())
print("脳腫瘍患者の平均ピクセルの分散:", group1.var())
```
